```python
import sqlite3
import pandas as pd
```

```python
dataset = pd.read_excel("data/salary_data_states.xlsx")
dataset
```

```python
print(dataset['JOB_TITLE_SUBGROUP'].unique())
```

```python
#remove unimportant job sub types

remove = ['attorney','teacher','assistant professor']

dataset = dataset[~dataset['JOB_TITLE_SUBGROUP'].isin(remove)]

dataset.head()
```

```python
print(dataset['WORK_STATE'].unique())
```

```python
#we dont care for guam, palau, northern mariana islands,Puerto Rico, virgin islands
remove_states = ['Guam', 'Palau', 'Northern Mariana Islands', 'Puerto Rico', 'Virgin Islands']

# Remove observations with specified work states
dataset = dataset[~dataset['WORK_STATE'].isin(remove_states)]

# Print the updated dataframe
dataset.head
```

```python
#Cleaning Data
print(dataset.isnull().sum())
```

```python
#Group by Job SubType of average wage
# Group data by job type and calculate mean salary for each group
mean_salary_by_job_type = dataset.groupby('JOB_TITLE_SUBGROUP')['PAID_WAGE_PER_YEAR'].median()


sorted_job_types = mean_salary_by_job_type.sort_values(ascending=False)

print(sorted_job_types)
```

```python
#Which companies have the highest salaries for those sub types
# Group data by job type and find the company that pays the highest salary for each group
highest_salary_by_job_type = dataset.groupby('JOB_TITLE_SUBGROUP').apply(lambda x: x.loc[x['PAID_WAGE_PER_YEAR'].idxmax()])

print(highest_salary_by_job_type[['EMPLOYER_NAME', 'PAID_WAGE_PER_YEAR']])
```

```python
#Changes with location of the job?
mean_salary_by_job_type_location = dataset.groupby('WORK_STATE')['PAID_WAGE_PER_YEAR'].median()


sorted_locations = mean_salary_by_job_type_location.sort_values(ascending=False)


print(sorted_locations)
```

```python
sorted_locations.head(10)
```

```python
#salary by jobsubgroup by state
salary_by_subtype_by_state = dataset.groupby(['WORK_STATE', 'JOB_TITLE_SUBGROUP'])['PAID_WAGE_PER_YEAR'].mean()
salary_by_subtype_by_state
```

```python
#sorted by all salaries
salary_by_subtype_by_state.sort_values(ascending=True)
```

```python
#Will the Answer Change if we take standard of living into account?
col_df = pd.read_csv('data/CostOfLiving2023.csv')
col_df
```

```python
#Cost of Living Index per State
cost_of_living = col_df[['state','costIndex']]
cost_of_living
```

```python
# Merge the two datasets based on state
merged_data = pd.merge(dataset, cost_of_living, left_on='WORK_STATE', right_on='state')
merged_data
```

To account for standard of living, we will divide the annual wage by the cost index by state. The adjusted salary represents the amount of money you would need to earn in a particular state to maintain the same wage with someone with a different standard of living in a different state. 

```python

merged_data['ADJUSTED_WAGE'] = merged_data['PAID_WAGE_PER_YEAR'] / merged_data['costIndex']

df1 = merged_data[['JOB_TITLE','JOB_TITLE_SUBGROUP','WORK_STATE','costIndex','PAID_WAGE_PER_YEAR','ADJUSTED_WAGE']]
df1
```
